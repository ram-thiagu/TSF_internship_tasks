```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt  
%matplotlib inline
```

* Let's import and view the dataset

```python
url = "http://bit.ly/w-data"
df = pd.read_csv(url)
df.head()
```

* Let's check the size of the dataset

```python
df.shape
```

* Let's visualize the dataset

```python
plt.scatter(x='Hours', y='Scores', data=df)
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Scored')  
plt.show()
```

* As per the above graph, we can say that there is a linear relationshp between hours studied and marks scored
* So, let's prepare the dataset for the Linear Regression model

```python
X=df.drop('Scores', axis=1)
y=df['Scores']
```

* Let's split the dataset into train and test.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

* Let's build and train the linear regression model

```python
from sklearn.linear_model import LinearRegression

model_LR=LinearRegression()
```

```python
model_LR.fit(X_train, y_train)
```

* Training completed

```python
line = model_LR.coef_*X+model_LR.intercept_

plt.scatter(X, y)
plt.title('Hours vs Percentage with the trained model line')
plt.xlabel('Number of Hours')
plt.ylabel('Number of Scores')
plt.plot(X, line)
plt.show()
```

* Let's Predict the values for X_test

```python
y_pred = model_LR.predict(X_test)

pred_vs_act_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
pred_vs_act_df
```

* Let's predict the scores for the 9.25 Hours

```python
hours = 9.25
pred = model_LR.predict([[hours]])
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(pred[0]))
```

* Let's evaluate the model's error rate

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
```
